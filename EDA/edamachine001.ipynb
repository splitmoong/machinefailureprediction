```python
import pandas as pd
import numpy as np
import scipy
```

```python
dataset = pd.read_csv("dataset/ai4i2020.csv")
columns_to_remove = ['UDI', 'Product ID']
dataset = dataset.drop(columns=columns_to_remove)
print(dataset)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=dataset, x='Machine failure', palette='cool')

plt.title("Distribution of Machine Failures")
plt.xlabel("Machine Failure (0 = No Failure, 1 = Failure)")
plt.ylabel("Count")
plt.show()

# Get the counts of machine fails
failure_counts = dataset['Machine failure'].value_counts()
print(failure_counts)
print(f"Pass (0): {failure_counts[0]}")
print(f"Fail (1): {failure_counts[1]}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_remove = ['Type']
dataset = dataset.drop(columns=columns_to_remove)
corr_matrix = dataset.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)

plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns = ['Air temperature [K]','Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]']

for feature in columns:
    plt.figure(figsize=(4, 3))  
    sns.boxplot(y=dataset[f'{feature}'], x=dataset['Machine failure'])
    plt.title(f'Box Plot of {feature} by DBU Class')
        
    lower_limit = dataset[f'{feature}'].quantile(0.05)
    upper_limit = dataset[f'{feature}'].quantile(0.95)
    plt.ylim(lower_limit, upper_limit)
        
    plt.show()
```

```python
simplified_dataset = dataset.drop(columns = ['Process temperature [K]', 'Rotational speed [rpm]'])

print(simplified_dataset)
```
